```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

#load dataset
kashti=sns.load_dataset("titanic")
kashti

#draw line plot
sns.barplot(x="class",y="fare",hue="sex", data=kashti, estimator=median)
plt.title("Kashti ka plot")
plt.show()
```
