# Total Fare and Survival Rate by Embarkation Port

```python
total_fare_and_survival = kashti.groupby('embark_town')[['fare', 'survived']].agg({
    'fare': 'sum',   # Total fare
    'survived': 'mean'  # Average survival rate
})
print(total_fare_and_survival)
```

# Maximum and Minimum Age by Class

```python
kashti.groupby('pclass')['age'].agg(['max', 'min'])
```
